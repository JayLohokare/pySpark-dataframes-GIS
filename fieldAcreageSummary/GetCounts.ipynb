```python
%pyspark
state_Ids =  field.select("field.state_id").distinct()
country_Ids = grower.select("grower.address.country_id").distinct()
```

```python
%pyspark
state_Ids.printSchema()
```

```python
%pyspark
realStates = state_Ids.join(states, state_Ids.state_id == states.state.id, 'left_outer')
```

```python
%pyspark
stateNames =  realStates.select('state.name').collect()
```

```python
%pyspark
string = ""
for i in stateNames :
    if i[0]!=None:
        string += (i[0])
        string += ', '
    
print string
```

```python
%pyspark
realCountriess = country_Ids.join(countries, country_Ids.country_id == countries.country.id, 'left_outer')
```

```python
%pyspark
countryNames =  realCountriess.select('country.alpha3').collect()

string = ""
for i in countryNames :
    if i[0]!=None:
        string += (i[0])
        string += ', '
    
print string
```

```python
%pyspark
countriesJoint = grower.join(realCountriess, realCountriess.country_id == grower.grower.address.country_id, 'left_outer')
countriesJoint.groupBy('country.alpha3').agg(F.count("country.alpha3").alias("count")).show()
```
